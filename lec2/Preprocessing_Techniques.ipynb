## Min-Max Scaling

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Generate a sample time series with some noise
time_series = np.sin(np.linspace(0, 2 * np.pi, 100)) * 2 + np.random.normal(0, 0.1, 100)

# Apply min-max scaling
scaler = MinMaxScaler()
time_series_scaled = scaler.fit_transform(time_series.reshape(-1, 1)).flatten()

# Plot both original and scaled time series on the same plot
plt.figure(figsize=(12, 6))

# Plot original time series
plt.plot(time_series, label='Original Time Series', color='blue')

# Plot min-max scaled time series
plt.plot(time_series_scaled, label='Min-Max Scaled Time Series', color='red')

# Add title and labels
plt.title('Original vs Min-Max Scaled Time Series')
plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.legend()

# Show the plot
plt.show()
```

## Standard Scaling

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Generate a sample time series with some noise
time_series = np.sin(np.linspace(0, 2 * np.pi, 100)) * 2 + np.random.normal(0, 0.1, 100)

# Apply standard scaling
scaler_standard = StandardScaler()
time_series_standard_scaled = scaler_standard.fit_transform(time_series.reshape(-1, 1)).flatten()

# Plot both original and standard scaled time series on the same plot
plt.figure(figsize=(12, 6))

# Plot original time series
plt.plot(time_series, label='Original Time Series', color='blue')

# Plot standard scaled time series
plt.plot(time_series_standard_scaled, label='Standard Scaled Time Series', color='red')

# Add title and labels
plt.title('Original vs Standard Scaled Time Series')
plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.legend()

# Show the plot
plt.show()
```

## L2 Normalizing

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer

# Generate a sample time series with some noise
time_series = np.sin(np.linspace(0, 2 * np.pi, 100)) * 1 + np.random.normal(0, 0.1, 100)

# Apply norm-based scaling (L2 normalization)
scaler_norm = Normalizer(norm='l2')
time_series_norm_scaled = scaler_norm.fit_transform(time_series.reshape(1, -1)).flatten()

# Plot both original and normalized time series on the same plot
plt.figure(figsize=(12, 6))

# Plot original time series
plt.plot(time_series, label='Original Time Series', color='blue')

# Plot norm-based scaled time series
plt.plot(time_series_norm_scaled, label='Norm-Based Scaled Time Series (L2)', color='purple')

# Add title and labels
plt.title('Original vs Norm-Based Scaled Time Series (L2)')
plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.legend()

# Show the plot
plt.show()
```

## Smoothing

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample time series
np.random.seed(42)
time_series = np.sin(np.linspace(0, 4 * np.pi, 100)) + np.random.normal(0, 0.1, 100)

# Moving Average functions
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

# Backward Moving Average
def backward_moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='full')[:len(data)]

# Forward Moving Average
def forward_moving_average(data, window_size):
    return np.convolve(data[::-1], np.ones(window_size) / window_size, mode='full')[:len(data)][::-1]

# Centered Moving Average
def centered_moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='same')

# Apply moving averages
window_size = 5
backward_avg = backward_moving_average(time_series, window_size)
forward_avg = forward_moving_average(time_series, window_size)
centered_avg = centered_moving_average(time_series, window_size)

# Plot the original time series and smoothed series
plt.figure(figsize=(18, 10))

# Original time series
plt.plot(time_series, label='Original Time Series', color='black', linestyle='dotted')

# Backward Moving Average
plt.plot(backward_avg, label='Backward Moving Average', color='blue')
plt.text(len(backward_avg) - 20, backward_avg[-20], 'Lagging Effect', fontsize=24, color='blue')

# Forward Moving Average
plt.plot(forward_avg, label='Forward Moving Average', color='green')
plt.text(20, forward_avg[20], 'Leading Effect', fontsize=24, color='green')

# Centered Moving Average
plt.plot(centered_avg, label='Centered Moving Average', color='orange')
plt.text(len(centered_avg) // 2, centered_avg[len(centered_avg) // 2], 'Balanced View', fontsize=24, color='orange')

# Titles and legend
plt.title('Moving Averages Comparison (Backward, Forward, Centered)')
plt.xlabel('Time Steps')
plt.ylabel('Values')
plt.legend()

# Show the plot
plt.tight_layout()
# plt.savefig('./img/time_series_smoothing.png', dpi=300)

plt.show()
```
