## Time Domain Features

### Data and Time

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a general sample time series with a linear trend and noise
np.random.seed(42)
time_steps = np.arange(0, 120)  # 10 years of monthly data
time_series = 50 + 2 * time_steps + np.random.normal(0, 10, size=len(time_steps))

# Generate multi-scale time series
yearly = time_series[::12]  # Yearly data from monthly data
monthly = time_series
weekly = np.interp(np.linspace(0, len(monthly)-1, len(monthly) * 4), np.arange(len(monthly)), monthly)  # Weekly data
daily = np.interp(np.linspace(0, len(monthly)-1, len(monthly) * 30), np.arange(len(monthly)), monthly)  # Daily data

# Plot the multi-scale time series
plt.figure(figsize=(16, 12))

# Yearly Time Series
plt.subplot(4, 1, 1)
plt.plot(yearly, label="Yearly Time Series", color='green')
plt.title("Yearly Time Series")
plt.xlabel("Years")
plt.ylabel("Values")
plt.legend()

# Monthly Time Series
plt.subplot(4, 1, 2)
plt.plot(monthly, label="Monthly Time Series", color='blue')
plt.title("Monthly Time Series")
plt.xlabel("Months")
plt.ylabel("Values")
plt.legend()

# Weekly Time Series
plt.subplot(4, 1, 3)
plt.plot(weekly, label="Weekly Time Series", color='orange')
plt.title("Weekly Time Series")
plt.xlabel("Weeks")
plt.ylabel("Values")
plt.legend()

# Daily Time Series
plt.subplot(4, 1, 4)
plt.plot(daily, label="Daily Time Series", color='red')
plt.title("Daily Time Series")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()

plt.tight_layout()
# plt.savefig("./img/data_and_time.png", dpi=300)
plt.show()
```

### Differencing

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic time series with a clear trend and seasonality
np.random.seed(42)
time = np.arange(300)
trend = 0.5 * time  # Linear trend
seasonality = 10 * np.sin(2 * np.pi * time / 20)  # Sinusoidal seasonality
noise = np.random.normal(0, 1, 300)  # Random noise
time_series = trend + seasonality + noise

# Perform first-order differencing
differenced_series_1 = np.diff(time_series, n=1)

# Perform second-order differencing
differenced_series_2 = np.diff(differenced_series_1, n=1)

# Plot the original time series, first-order, and second-order differenced series
plt.figure(figsize=(12, 12))

# Original time series
plt.subplot(3, 1, 1)
plt.plot(time, time_series, label='Original Time Series', color='blue')
plt.title('Original Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()

# First-order differenced time series
plt.subplot(3, 1, 2)
plt.plot(time[1:], differenced_series_1, label='First-Order Differenced Series', color='orange')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.title('First-Order Differenced Time Series')
plt.xlabel('Time')
plt.ylabel('Differenced Value')
plt.legend()

# Second-order differenced time series
plt.subplot(3, 1, 3)
plt.plot(time[2:], differenced_series_2, label='Second-Order Differenced Series', color='green')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.title('Second-Order Differenced Time Series')
plt.xlabel('Time')
plt.ylabel('Differenced Value')
plt.legend()

plt.tight_layout()
plt.show()
```
