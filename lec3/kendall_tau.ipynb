```python
# Kendall's tau for time series analysis
import numpy as np
from scipy.stats import kendalltau
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Time Series Data
temperature = [27.25, 28.0, 28.92, 26.5, 29.1, 29.5, 29.89, 31.0, 29.29, 30.62, 31.38, 29.75, 30.5, 30.93, 29.23, 31.22, 27.0, 25.62, 27.12, 27.86]
pressure = [1009.75, 1008.88, 1010.58, 1009.88, 1010.2, 1009.5, 1009.0, 1007.12, 1007.14, 1007.5, 1005.0, 1004.25, 1004.25, 1007.2, 1005.0, 1002.89, 1007.38, 1010.38, 1010.62, 1008.57]

# Normalize
temperature = (temperature - np.mean(temperature)) / np.std(temperature)
pressure = (pressure - np.mean(pressure)) / np.std(pressure)

sns.set()
plt.figure(figsize=(8, 3))
plt.xticks(range(0, 20), range(0, 20))
plt.plot(temperature, label='Temperature (z-Normalized)', color='red', marker='o')
plt.plot(pressure, label='Pressure (z-Normalized)', color='blue', marker='x')
plt.legend()
```

```python
# Calculate Kendall's Tau
tau, p_value = kendalltau(temperature, pressure)

print(f"Kendall's Tau: {tau}")
print(f"P-value: {p_value}")

# Calculate Pearson
corr = np.corrcoef(temperature, pressure)[0, 1]
print(f"Pearson's Correlation: {corr}")
```
