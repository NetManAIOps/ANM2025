```python
# Kendall's tau for time series analysis
import numpy as np
from scipy.stats import kendalltau
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Time Series Data
temperature = np.array([22, 24, 23, 21, 25])
precipitation = np.array([30, 35, 35, 33, 40])

sns.set()
plt.figure(figsize=(8, 3))
plt.plot(temperature, label='Temperature', color='red', marker='o')
plt.plot(precipitation, label='Precipitation', color='blue', marker='x')
plt.legend()
plt.savefig('test.pdf')
```

```python
# Calculate Kendall's Tau
tau, p_value = kendalltau(temperature, precipitation)

print(f"Kendall's Tau: {tau}")
print(f"P-value: {p_value}")
```
