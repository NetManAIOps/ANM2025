```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t = np.arange(50)
```

```python
# Generate time series with strong positive correlation
x1 = t * 0.1 + np.random.normal(0, 0.1, 50)
x2 = t * 1 + np.random.normal(0, 1, 50)

# Calculate correlation
correlation_matrix = np.corrcoef(x1, x2)
pearson_correlation = correlation_matrix[0, 1]

# Plot
sns.set()
plt.figure(figsize=(15, 4))
plt.subplot(121)
plt.title('x1')
plt.plot(t, x1)
plt.subplot(122)
plt.title('x2')
plt.plot(t, x2)
plt.suptitle(f"Pearson Correlation: {pearson_correlation:.3f}")
```

```python
# Time series with strong negative correlation
x1 = t * -0.1 + np.random.normal(0, 0.1, 50)
x2 = t * 1 + np.random.normal(0, 1, 50)

# Calculate correlation
correlation_matrix = np.corrcoef(x1, x2)
pearson_correlation = correlation_matrix[0, 1]

# Plot
sns.set()
plt.figure(figsize=(15, 4))
plt.subplot(121)
plt.title('x1')
plt.plot(t, x1)
plt.subplot(122)
plt.title('x2')
plt.plot(t, x2)
plt.suptitle(f"Pearson Correlation: {pearson_correlation:.3f}")
```

```python
# Time series with almost no correlation
x1 = t * -0.1 + np.random.normal(0, 0.1, 50)
x2 = np.random.normal(0, 1, 50)

# Calculate correlation
correlation_matrix = np.corrcoef(x1, x2)
pearson_correlation = correlation_matrix[0, 1]

# Plot
sns.set()
plt.figure(figsize=(15, 4))
plt.subplot(121)
plt.title('x1')
plt.plot(t, x1)
plt.subplot(122)
plt.title('x2')
plt.plot(t, x2)
plt.suptitle(f"Pearson Correlation: {pearson_correlation:.3f}")
```

```python
# CORT coefficient
def cort(x1: np.ndarray, x2: np.ndarray, eps: float=1e-7) -> float:
    d_x1 = x1[1:] - x1[:-1]
    d_x2 = x2[1:] - x2[:-1]

    result = np.sum(d_x1 * d_x2) / np.sqrt(np.sum(d_x1 ** 2) * np.sum(d_x2 ** 2) + eps)
    return result
```

```python
# Generate time series with strong positive CORT correlation
x1 = t * 0.1
x2 = t * 1.5
x1[20:] += 1.0
x2[20:] += 15.0

# Calculate correlation
cort_correlation = cort(x1, x2)

# Plot
sns.set()
plt.figure(figsize=(15, 5))
plt.subplot(211)
plt.title('x1')
plt.plot(t, x1)
plt.subplot(212)
plt.title('x2')
plt.plot(t, x2)
plt.suptitle(f"CORT Correlation: {cort_correlation:.3f}")
```

```python
# Generate time series with strong negative CORT correlation
x1 = t * -0.1
x2 = t * 1.5
x1[20:] -= 1.0
x2[20:] += 15.0

# Calculate correlation
cort_correlation = cort(x1, x2)

# Plot
sns.set()
plt.figure(figsize=(15, 5))
plt.subplot(211)
plt.title('x1')
plt.plot(t, x1)
plt.subplot(212)
plt.title('x2')
plt.plot(t, x2)
plt.suptitle(f"CORT Correlation: {cort_correlation:.3f}")
```

```python
# Compare CORT with Pearson
x1 = t * 0.1
x2 = t * 1.5
x1[20:] -= 1.0
x2[20:] += 15.0

# Calculate correlation
cort_correlation = cort(x1, x2)
pearson_correlation = np.corrcoef(x1, x2)[0, 1]

# Plot
sns.set()
plt.figure(figsize=(15, 5))
plt.subplot(211)
plt.title('x1')
plt.plot(t, x1)
plt.subplot(212)
plt.title('x2')
plt.plot(t, x2)
plt.suptitle(f"CORT: {cort_correlation:.3f}; Pearson: {pearson_correlation:.3f}")
```
