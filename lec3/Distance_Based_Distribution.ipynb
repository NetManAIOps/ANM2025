## Distance Based on Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance

# Generate two time series
np.random.seed(0)
time = np.linspace(0, 10, 1000)
time_series1 = np.sin(time) + np.random.normal(0, 0.5, 1000)  # Sine wave with noise
time_series2 = np.sin(time + 1) + np.random.normal(0, 0.5, 1000)  # Sine wave with noise, phase shift

# Calculate the Wasserstein distance between the time series
distance = wasserstein_distance(time_series1, time_series2)

# Plot the two time series and their distribution plots
fig, ax = plt.subplots(2, 1, figsize=(10, 12))

# Plot the time series
ax[0].plot(time, time_series1, label='Time Series 1', color='skyblue')
ax[0].plot(time, time_series2, label='Time Series 2', color='salmon')
ax[0].set_title('Time Series 1 and Time Series 2', fontsize=14)
ax[0].set_xlabel('Time', fontsize=12)
ax[0].set_ylabel('Value', fontsize=12)
ax[0].legend()

# Plot the empirical distributions of the time series
ax[1].hist(time_series1, bins=30, alpha=0.5, label='Empirical Distribution 1', color='skyblue')
ax[1].hist(time_series2, bins=30, alpha=0.5, label='Empirical Distribution 2', color='salmon')
ax[1].set_title(f'Empirical Distributions and Their Wasserstein Distance\n(Wasserstein Distance: {distance:.2f})', fontsize=14)
ax[1].set_xlabel('Value', fontsize=12)
ax[1].set_ylabel('Frequency', fontsize=12)
ax[1].legend()

# Display the plot
plt.tight_layout()
plt.show()
```
